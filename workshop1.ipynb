```python
# Vector addition using numpy
import numpy as np
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])
v1 + v2
print(v1 + v2)
0.5 * v2
```

```python
# function to add two vectors
def add(v1, v2):
    if len(v1) == len(v2):
        return v1 + v2
    else:
        return "Vectors are not the same length"
```

```python
import matplotlib.colors as mc
import matplotlib.pyplot as plt

red = mc.to_rgb([1, 0, 0])
green = mc.to_rgb([0, 1, 0])
blue = mc.to_rgb([0, 0, 1])

plt.plot(2, 4, "o")
plt.plot(2, 4, "s")
plt.plot(2, 4, "ro")
plt.plot(2, 4, "o", color="r", markersize=50)
plt.plot(2, 4, "o", color=blue, markersize=50)
```

```python
def color_mix(c1, c2):
    return [(c1[i] + c2[i]) / 2 for i in range(3)]

mycolor = color_mix(green, red)
plt.plot(2, 4, "o", color=mycolor, markersize=50)
```

```python
def color_mix2(c1, c2):
    t1 = np.array(c1)
    t2 = np.array(c2)
    return mc.to_rgb(0.5*t1 + 0.5*t2)

mycolor = color_mix2(red, green)
plt.plot(2, 4, "o", color=mycolor, markersize=50)
```
