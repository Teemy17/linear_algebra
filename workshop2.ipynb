```python
import numpy as np

v0 = np.array([0, 0])  # vector 0 or the origin
v1 = np.array([1, 3])
v2 = np.array([4, 2])
v3 = np.array([1, 3, 5])

# dot product
np.dot(v1, v2)
v1.dot(v2)
v1 @ v2
```

```python
# using the definition of length
np.sqrt(v1.dot(v1))
np.sqrt(v1 @ v1)
(v3 @ v3) ** 0.5
# using linear algebra library
np.linalg.norm(v1)
np.linalg.norm(v3)
```

```python
# find the unit vector u of a given vector v
u1 = v1 / np.sqrt(v1 @ v1)
print(u1)
u2 = v2 / np.sqrt(v2 @ v2)
print(u2)


def v_unit(v):
    return v / np.sqrt(v @ v)


print(v_unit(v1))
print(v_unit(v3))
```

```python
def v_angle(v1, v2):
    u1 = v1 / np.sqrt(v1 @ v1)
    u2 = v2 / np.sqrt(v2 @ v2)
    angle = np.arccos(u1 @ u2) * 180 / np.pi
    return angle


v_angle(v1, v2)
```

```python
# Vector plot
import matplotlib.pyplot as plt


def vplot(v):
    plt.plot([0, v[0]], [0, v[1]])


vplot(v1)
vplot(v2)
```

```python
plt.plot([0, 2], [0, 3], "r--", label="u")
plt.plot([0, 4], [0, 1], "g-.", label="v")
plt.legend()
```

```python
# Line style, Color, LAbel and Legend
def vplot(v, style, label):
    plt.plot([0, v[0]], [0, v[1]], style, label=label)
    plt.legend()


vplot(v1, "--", "v1")
vplot(v2, "-.", "v2")
vplot(v1 + v2, "-", "v1+v2")
```

```python
# Vector Combinations
def vcomb(u, v, a):
    if a < 0 or a > 1:
        return "error: a must be between 0 and 1"
    else:
        return (1 - a) * u + a * v


print(vcomb(v1, v2, 0))
print(vcomb(v1, v2, 1))
print(vcomb(v1, v2, 0.5))
```

```python
vplot(v1,'--', 'u')
vplot(v2,'-.', 'v')

vplot(vcomb(v1, v2, 0.1), '-', 'v01')
vplot(vcomb(v1, v2, 0.3), '-', 'v03')
vplot(vcomb(v1, v2, 0.5), '-', 'v05')
vplot(vcomb(v1, v2, 0.7), '-', 'v07')
vplot(vcomb(v1, v2, 0.9), '-', 'v09')
```
